```python
# data cleaning

# renaming the df column names to lowercase

death_df.columns = map(str.lower, death_df.columns)
confirmed_df.columns = map(str.lower, confirmed_df.columns)
recovered_df.columns = map(str.lower, recovered_df.columns)
```

```python
# rename province/state to 'state', country/region to 'country'
confirmed_df = confirmed_df.rename(columns={'province/state': 'state', 'country/region': 'country'})
death_df = death_df.rename(columns={'province/state': 'state', 'country/region': 'country'})
recovered_df = recovered_df.rename(columns={'province/state': 'state', 'country/region': 'country'})
```

```python
recovered_df.drop('state', axis='columns', inplace=True)
confirmed_df.drop('state', axis='columns', inplace=True)
death_df.drop('state', axis='columns', inplace=True)
```

```python
confirmed_df = confirmed_df.rename(columns={'11/8/2020': 'confirmed'})
death_df = death_df.rename(columns={'11/8/2020': 'death'})
recovered_df = recovered_df.rename(columns={'11/8/2020': 'recovered'})
```

```python
recovered_df_dropped = recovered_df.filter(['country','lat', 'long', 'recovered'])
```

```python
confirmed_df_dropped = confirmed_df.filter(['country','lat', 'long', 'confirmed'])
```

```python
death_df_dropped = death_df.filter(['country','lat', 'long', 'death'])
```

```python
recovered_df_dropped = recovered_df_dropped.groupby('country').agg({'lat':'first', 'long': 'first', 'recovered': 'sum'}).reset_index()
```

```python
confirmed_df_dropped = confirmed_df_dropped.groupby('country').agg({'lat':'first', 'long': 'first', 'confirmed': 'sum'}).reset_index()
```

```python
death_df_dropped = death_df_dropped.groupby('country').agg({'lat':'first', 'long': 'first', 'death': 'sum'}).reset_index()
```

```python
countries = pd.DataFrame()
```

```python
countries = confirmed_df_dropped.filter(['country','lat', 'long', 'confirmed'])
```

```python
selected_columns = recovered_df_dropped[["recovered"]]
```

```python
countries = countries.join(selected_columns)
```

```python
selected_columns = death_df_dropped[["death"]]
```

```python
countries = countries.join(selected_columns)
```

```python
countries['active'] = 0
```

```python
countries['active'] = countries['confirmed']-countries['recovered']-countries['death']
```

```python
confirmed_total = int(countries['confirmed'].sum())
deaths_total = int(countries['death'].sum())
recovered_total = int(countries['recovered'].sum())
active_total = int(countries['active'].sum())
```

```python
import plotly.graph_objects as go
from ipywidgets import interact, interactive, fixed, interact_manual

def highlight_col(x):
    r = 'background-color: red'
    y = 'background-color: lightgreen'
    g = 'background-color: lightblue'
    c = 'background-color: orange'
    df1 = pd.DataFrame('', index=x.index, columns=x.columns)
    
    df1.iloc[:, 3] = c
    df1.iloc[:, 4] = y
    df1.iloc[:, 5] = r
    df1.iloc[:, 6] = g
    
    return df1

def show_latest_cases(n=10):
    n = int(n)
    return sorted_country_df.sort_values('confirmed', ascending= False).head(n).style.apply(highlight_col, axis=None)

interact(show_latest_cases, n=(1,200,1))
```

```python
def show_latest_cases_by_country(x):
    # return sorted_country_df.country[str(x)].head(1).style.apply(highlight_col, axis=None)
    return sorted_country_df.loc[sorted_country_df['country'] == x].style.apply(highlight_col, axis=None)
    
interact(show_latest_cases_by_country, x=sorted_country_df.country)
```

```python
import plotly.express as px
```

```python
def bubble_chart_combined(val, n=10):

    fig = px.scatter(countries.sort_values(str.lower(val), ascending= False).head(n), 
                     x="country", y=str.lower(val), size=str.lower(val), color="country",
                     hover_name="country", size_max=60)
    fig.update_layout(
    title=str(n) +" Countries by "+ str(val) + " Cases",
    xaxis_title= "Countries",
    yaxis_title= str(val) + " Cases",
    width = 700
    )
    fig.show();

interact(bubble_chart_combined, val=['Confirmed', 'Death', 'Recovered', 'Active'], n=(1,200,1))
```

```python
def plot_cases_of_a_country(country):
    labels = ['confirmed', 'deaths']
    colors = ['blue', 'red']
    mode_size = [6, 8]
    line_size = [4, 5]
    
    df_list = [confirmed_df, death_df]
    
    fig = go.Figure();
    
    for i, df in enumerate(df_list):
        if country == 'World' or country == 'world':
            x_data = np.array(list(df.iloc[:, :].columns))
            y_data = np.sum(np.asarray(df.iloc[:,4:]),axis = 0)
            
        else:    
            x_data = np.array(list(df.iloc[:, :].columns))
            y_data = np.sum(np.asarray(df[df['country'] == country].iloc[:,4:]),axis = 0)
            
        fig.add_trace(go.Scatter(x=x_data, y=y_data, mode='lines+markers',
        name=labels[i],
        line=dict(color=colors[i], width=line_size[i]),
        connectgaps=True,
        text = "Total " + str(labels[i]) +": "+ str(y_data[-1])
        ));
    
    fig.update_layout(
        title="COVID 19 cases of " + country,
        xaxis_title='Date',
        yaxis_title='No. of Confirmed Cases',
        margin=dict(l=20, r=20, t=40, b=20),
        paper_bgcolor="lightgrey",
        width = 800,
        
    );
    
    fig.update_yaxes(type="linear")
    fig.show();
```

```python
world_countries = ['World']
world_countries.extend(countries.country)
interact(plot_cases_of_a_country, country=world_countries)
```

```python
world_map = folium.Map(location=[11,0], tiles="cartodbpositron", zoom_start=2, max_zoom = 6, min_zoom = 2)


for i in range(0,len(confirmed_df)):
    folium.Circle(
        location=[confirmed_df.iloc[i]['lat'], confirmed_df.iloc[i]['long']],
        fill=True,
        radius=(int((np.log(confirmed_df.iloc[i,-1]+1.00001)))+0.2)*50000,
        color='red',
        fill_color='indigo',
        tooltip = "<div style='margin: 0; background-color: black; color: white;'>"+
                    "<h4 style='text-align:center;font-weight: bold'>"+confirmed_df.iloc[i]['country'] + "</h4>"
                    "<hr style='margin:10px;color: white;'>"+
                    "<ul style='color: white;;list-style-type:circle;align-item:left;padding-left:20px;padding-right:20px'>"+
                        "<li>Confirmed: "+str(confirmed_df.iloc[i,-1])+"</li>"+
                        "<li>Deaths:   "+str(death_df.iloc[i,-1])+"</li>"+
                        "<li>Death Rate: "+ str(np.round(death_df.iloc[i,-1]/(confirmed_df.iloc[i,-1]+1.00001)*100,2))+ "</li>"+
                    "</ul></div>",
        ).add_to(world_map)

world_map
```
